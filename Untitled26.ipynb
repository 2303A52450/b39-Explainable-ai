```python
#!/usr/bin/env python3
"""
Full UTI pipeline:
- Missing-value handling (median/mode)
- Encoding categorical features
- Scaling (StandardScaler or MinMaxScaler)
- Imbalance handling (SMOTE preferred, else random oversample)
- Correlation analysis (heatmap) and removal (|corr| > 0.85)
- SHAP explainability (summary, bar, dependence top-3) with fallback to permutation importance
- Visualization: histogram, boxplot, scatter, bar, pie, heatmap, pairplot (if small), violin
- ML: LogisticRegression, DecisionTree, KNN, GaussianNB, SVM
- Ensembles: RandomForest, ExtraTrees, AdaBoost, GradientBoosting, Voting, Stacking
- Deep Learning: Keras MLP if available else sklearn MLP
- Metrics saved: accuracy, precision, recall, f1, auc; confusion matrices and ROC plots saved

Author: ChatGPT (GPT-5 Thinking mini)
"""

import os
import sys
import warnings
warnings.filterwarnings("ignore")
import json
import pickle
from datetime import datetime
import subprocess

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# sklearn
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             confusion_matrix, roc_auc_score, roc_curve, classification_report)
from sklearn.utils import resample
from sklearn.inspection import permutation_importance

# classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import (RandomForestClassifier, ExtraTreesClassifier,
                              AdaBoostClassifier, GradientBoostingClassifier,
                              VotingClassifier, StackingClassifier)

# optional libs - graceful fallback
HAVE_IMBLEARN = True
try:
    from imblearn.over_sampling import SMOTE
except Exception:
    HAVE_IMBLEARN = False

HAVE_SHAP = True
try:
    import shap
except Exception:
    HAVE_SHAP = False
    # attempt install (best-effort; may fail in restricted env)
    try:
        print("Attempting to pip install shap (may take a while)...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "shap"], stdout=subprocess.DEVNULL)
        import shap
        HAVE_SHAP = True
        print("Installed shap.")
    except Exception:
        HAVE_SHAP = False
        print("Could not install shap; will use permutation importance as fallback.")

HAVE_XGB = True
try:
    import xgboost as xgb
except Exception:
    HAVE_XGB = False

HAVE_LGB = True
try:
    import lightgbm as lgb
except Exception:
    HAVE_LGB = False

HAVE_CAT = True
try:
    import catboost as cb
except Exception:
    HAVE_CAT = False

HAVE_KERAS = True
try:
    from tensorflow import keras
    from tensorflow.keras import layers
except Exception:
    HAVE_KERAS = False

# -----------------------------
# User parameters (edit here)
# -----------------------------
DATA_PATH = "UTI.csv"   # path to dataset
TARGET_NAME = "uti"               # target column name as user said
SCALER_CHOICE = "standard"        # "standard" or "minmax"
CORR_THRESHOLD = 0.85
TOP_K_SHAP = 10
TEST_SIZE = 0.2
RANDOM_STATE = 42
OUT_DIR = "outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# -----------------------------
# Utility functions
# -----------------------------
def save_fig(fig, fname):
    p = os.path.join(OUT_DIR, fname)
    fig.tight_layout()
    fig.savefig(p, dpi=200, bbox_inches='tight')
    plt.close(fig)
    return p

def ensure_binary(y):
    return len(np.unique(y)) == 2

def metrics_summary(y_true, y_pred, y_prob=None):
    avg = 'binary' if ensure_binary(y_true) else 'macro'
    out = {
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "precision": float(precision_score(y_true, y_pred, average=avg, zero_division=0)),
        "recall": float(recall_score(y_true, y_pred, average=avg, zero_division=0)),
        "f1": float(f1_score(y_true, y_pred, average=avg, zero_division=0))
    }
    if (y_prob is not None) and ensure_binary(y_true):
        try:
            out["auc"] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            out["auc"] = None
    else:
        out["auc"] = None
    return out

# -----------------------------
# Load dataset
# -----------------------------
print("Loading dataset:", DATA_PATH)
if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"Dataset not found at {DATA_PATH}. Place UTI.csv there or edit DATA_PATH.")

df = pd.read_csv(DATA_PATH)
print("Dataset shape:", df.shape)
print("Columns:", df.columns.tolist())

if TARGET_NAME not in df.columns:
    # try common names fallback
    possible = [c for c in df.columns if c.lower() in ('target','label','class','diagnosis','uti','infection','result')]
    if len(possible) > 0:
        TARGET_NAME = possible[0]
        print("Target not found, using fallback target:", TARGET_NAME)
    else:
        TARGET_NAME = df.columns[-1]
        print("Target not found; using last column as target:", TARGET_NAME)

# basic EDA prints
print("\nFirst rows:\n", df.head())
print("\nDtypes:\n", df.dtypes)
print("\nMissing per column:\n", df.isnull().sum())

# -----------------------------
# 1) Missing value handling
# -----------------------------
# Drop columns with extremely high missing fraction (>60%)
col_missing_frac = df.isnull().mean()
cols_drop = col_missing_frac[col_missing_frac > 0.6].index.tolist()
if cols_drop:
    print("Dropping columns with >60% missing:", cols_drop)
    df = df.drop(columns=cols_drop)

# Drop rows that are entirely NaN
df = df.dropna(how='all').reset_index(drop=True)

# Separate X and y
X = df.drop(columns=[TARGET_NAME]).copy()
y = df[TARGET_NAME].copy()

# Impute
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X.select_dtypes(exclude=[np.number]).columns.tolist()
print("Numeric columns:", num_cols)
print("Categorical columns:", cat_cols)

if num_cols:
    num_imp = SimpleImputer(strategy='median')
    X[num_cols] = num_imp.fit_transform(X[num_cols])
if cat_cols:
    cat_imp = SimpleImputer(strategy='most_frequent')
    X[cat_cols] = cat_imp.fit_transform(X[cat_cols])
    # encode categorical
    for c in cat_cols:
        X[c] = LabelEncoder().fit_transform(X[c].astype(str))

# encode target if necessary
if y.dtype == object or y.dtype.name == 'category':
    y = LabelEncoder().fit_transform(y.astype(str))
else:
    try:
        # safe cast to int if 0/1-like
        if set(np.unique(y.astype(int))) <= set([0,1]):
            y = y.astype(int)
    except Exception:
        pass

print("Missing after imputation:", X.isnull().sum().sum())

# -----------------------------
# 2) Scaling (choose)
# -----------------------------
print("Scaling method:", SCALER_CHOICE)
if SCALER_CHOICE.lower() == "minmax":
    scaler = MinMaxScaler()
else:
    scaler = StandardScaler()

X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_scaled.to_csv(os.path.join(OUT_DIR, "X_scaled.csv"), index=False)

# -----------------------------
# 3) Imbalance handling (SMOTE preferred)
# -----------------------------
print("Checking class distribution...")
vals, cnts = np.unique(y, return_counts=True)
print("Before balancing:", dict(zip(vals, cnts)))

X_bal = X_scaled.copy()
y_bal = y.copy()

if len(cnts) > 1 and cnts.max()/cnts.min() > 1.5:
    print("Imbalance detected.")
    if HAVE_IMBLEARN:
        try:
            sm = SMOTE(random_state=RANDOM_STATE)
            X_bal, y_bal = sm.fit_resample(X_bal, y_bal)
            print("SMOTE applied.")
        except Exception as e:
            print("SMOTE failed, falling back to random oversample:", e)
            HAVE_IMBLEARN = False
    if not HAVE_IMBLEARN:
        # random oversample minority to match majority
        df_comb = pd.concat([X_bal, pd.Series(y_bal, name=TARGET_NAME)], axis=1)
        maj = df_comb[TARGET_NAME].value_counts().max()
        frames = []
        for cls, grp in df_comb.groupby(TARGET_NAME):
            if len(grp) < maj:
                frames.append(resample(grp, replace=True, n_samples=maj, random_state=RANDOM_STATE))
            else:
                frames.append(grp)
        df_bal = pd.concat(frames).sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)
        y_bal = df_bal[TARGET_NAME].values
        X_bal = df_bal.drop(columns=[TARGET_NAME])
        print("Random oversampling applied.")
else:
    print("No major imbalance detected; skipping oversampling.")

print("After balancing class distribution:", dict(pd.Series(y_bal).value_counts()))

# -----------------------------
# 4) Correlation analysis and removal
# -----------------------------
print("Computing correlation matrix and removing highly correlated features (>|{:.2f}|)...".format(CORR_THRESHOLD))
corr = X_bal.corr().abs()
corr.to_csv(os.path.join(OUT_DIR, "correlation_matrix.csv"), index=True)

# Heatmap plot
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
heatmap_path = os.path.join(OUT_DIR, "correlation_heatmap.png")
plt.savefig(heatmap_path, dpi=200, bbox_inches='tight')
plt.show(); plt.clf()
print("Saved heatmap to:", heatmap_path)

# remove features with high pairwise correlation
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > CORR_THRESHOLD)]
print("Dropping due to high correlation:", to_drop)
X_uncorr = X_bal.drop(columns=to_drop, errors='ignore')
X_uncorr.to_csv(os.path.join(OUT_DIR, "X_uncorr.csv"), index=False)

# -----------------------------
# 5) Feature importance via SHAP (or permutation fallback)
# -----------------------------
print("Computing feature importance via SHAP (fallback to permutation importance if SHAP unavailable)...")
selected_features = []
shap_available = HAVE_SHAP

if shap_available:
    try:
        rf_shap = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=1)
        rf_shap.fit(X_uncorr, y_bal)
        explainer = shap.TreeExplainer(rf_shap)
        shap_values = explainer.shap_values(X_uncorr, check_additivity=False)
        # choose positive class shap array when binary
        if isinstance(shap_values, list) and len(shap_values) == 2:
            shap_arr = shap_values[1]
        else:
            shap_arr = shap_values
        shap_mean_abs = np.abs(shap_arr).mean(axis=0)
        shap_ser = pd.Series(shap_mean_abs, index=X_uncorr.columns).sort_values(ascending=False)
        shap_ser.to_csv(os.path.join(OUT_DIR, "shap_mean_abs.csv"))
        selected_features = list(shap_ser.head(TOP_K_SHAP).index)
        print("Top SHAP features:", selected_features)
    except Exception as e:
        print("SHAP failed:", e)
        shap_available = False

if not shap_available:
    print("Using permutation importance fallback.")
    rf_perm = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=1)
    rf_perm.fit(X_uncorr, y_bal)
    perm = permutation_importance(rf_perm, X_uncorr, y_bal, n_repeats=10, random_state=RANDOM_STATE, n_jobs=1)
    perm_ser = pd.Series(perm.importances_mean, index=X_uncorr.columns).sort_values(ascending=False)
    perm_ser.to_csv(os.path.join(OUT_DIR, "permutation_importances.csv"))
    selected_features = list(perm_ser.head(TOP_K_SHAP).index)
    print("Top permutation features:", selected_features)

# choose final features (top K) - intersection with X_uncorr columns
final_features = [f for f in selected_features if f in X_uncorr.columns]
if len(final_features) == 0:
    raise RuntimeError("No features selected — check dataset or TOP_K_SHAP.")

print("Final selected features:", final_features)
pd.Series(final_features).to_csv(os.path.join(OUT_DIR, "selected_features.csv"), index=False)
X_final = X_uncorr[final_features].copy()
X_final.to_csv(os.path.join(OUT_DIR, "X_final.csv"), index=False)

# -----------------------------
# 6) SHAP visualizations (advanced): summary, bar, dependence top3
# -----------------------------
if shap_available and 'shap_arr' in locals():
    try:
        print("Creating SHAP summary (beeswarm) plot...")
        shap.summary_plot(shap_arr, X_uncorr, show=False)
        plt.savefig(os.path.join(OUT_DIR, "shap_summary.png"), dpi=200, bbox_inches='tight'); plt.clf()
        print("Saved shap_summary.png")

        print("Creating SHAP bar plot (mean abs)...")
        shap.summary_plot(shap_arr, X_uncorr, plot_type="bar", show=False)
        plt.savefig(os.path.join(OUT_DIR, "shap_bar.png"), dpi=200, bbox_inches='tight'); plt.clf()
        print("Saved shap_bar.png")

        # Dependence plots for top 3 features (if available)
        for feat in final_features[:3]:
            try:
                shap.dependence_plot(feat, shap_arr, X_uncorr, show=False)
                plt.savefig(os.path.join(OUT_DIR, f"shap_dependence_{feat}.png"), dpi=200, bbox_inches='tight'); plt.clf()
                print("Saved shap_dependence_{}.png".format(feat))
            except Exception as e:
                print("Dependence plot failed for", feat, e)
    except Exception as e:
        print("SHAP plotting failed:", e)
else:
    print("SHAP not available; skip SHAP plots (see permutation_importances.csv).")

# -----------------------------
# 7) Visualizations requested: histogram, boxplot, scatter, bar chart (class counts), pie, pairplot (if not too large), violin
# -----------------------------
print("Generating visualizations (saved to outputs/) and showing key ones...")

# Histogram for first selected feature
f1 = final_features[0]
plt.figure(figsize=(6,4)); sns.histplot(X_final[f1], bins=20, kde=True); plt.title(f"Histogram: {f1}")
plt.savefig(os.path.join(OUT_DIR, f"hist_{f1}.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()

# Boxplot for second selected feature (if exists)
if len(final_features) > 1:
    f2 = final_features[1]
    plt.figure(figsize=(6,4)); sns.boxplot(x=X_final[f2]); plt.title(f"Boxplot: {f2}")
    plt.savefig(os.path.join(OUT_DIR, f"box_{f2}.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()

# Scatter of top2
if len(final_features) > 1:
    plt.figure(figsize=(6,4)); plt.scatter(X_final[final_features[0]], X_final[final_features[1]], alpha=0.7)
    plt.xlabel(final_features[0]); plt.ylabel(final_features[1]); plt.title("Scatter: top2 features")
    plt.savefig(os.path.join(OUT_DIR, "scatter_top2.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()

# Bar chart of class counts
plt.figure(figsize=(6,4)); vals, counts = np.unique(y_bal, return_counts=True); plt.bar([str(v) for v in vals], counts)
plt.title("Class counts (after balancing)"); plt.savefig(os.path.join(OUT_DIR, "bar_class_counts.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()

# Pie chart of class distribution
plt.figure(figsize=(5,5)); plt.pie(counts, labels=[str(v) for v in vals], autopct="%1.1f%%"); plt.title("Class distribution"); plt.savefig(os.path.join(OUT_DIR, "pie_class_dist.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()

# Pairplot (only if dataset small to avoid heavy plotting)
if X_final.shape[0] <= 500 and X_final.shape[1] <= 6:
    try:
        sns.pairplot(pd.concat([X_final, pd.Series(y_bal, name=TARGET_NAME)], axis=1), hue=TARGET_NAME, diag_kind="kde")
        plt.savefig(os.path.join(OUT_DIR, "pairplot_selected.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()
    except Exception as e:
        print("Pairplot failed:", e)
else:
    print("Pairplot skipped (too large).")

# Violin plot for first few features
plt.figure(figsize=(8,6))
sns.violinplot(data=X_final.iloc[:, :min(6, X_final.shape[1])], inner="box")
plt.title("Violin plot (first up to 6 selected features)")
plt.savefig(os.path.join(OUT_DIR, "violin_selected.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()

# -----------------------------
# 8) Train/test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(X_final, y_bal, test_size=TEST_SIZE, stratify=y_bal if len(np.unique(y_bal))>1 else None, random_state=RANDOM_STATE)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# -----------------------------
# 9) Train classical ML models and evaluate
# -----------------------------
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    "DecisionTree": DecisionTreeClassifier(random_state=RANDOM_STATE),
    "KNN": KNeighborsClassifier(),
    "GaussianNB": GaussianNB(),
    "SVM": SVC(probability=True, random_state=RANDOM_STATE)
}
results = {}
fitted = {}

for name, mdl in models.items():
    try:
        mdl.fit(X_train, y_train)
        fitted[name] = mdl
        y_pred = mdl.predict(X_test)
        y_prob = None
        if hasattr(mdl, "predict_proba") and ensure_binary(y_train):
            try: y_prob = mdl.predict_proba(X_test)[:,1]
            except: y_prob = None
        elif hasattr(mdl, "decision_function"):
            try: y_prob = mdl.decision_function(X_test)
            except: y_prob = None
        res = metrics_summary(y_test, y_pred, y_prob)
        results[name] = res
        print(f"\n[{name}] metrics:", res)
        print(classification_report(y_test, y_pred, zero_division=0))
        # Confusion matrix plot
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(4,3)); sns.heatmap(cm, annot=True, fmt="d"); plt.title(f"Confusion Matrix - {name}")
        plt.savefig(os.path.join(OUT_DIR, f"confusion_{name}.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()
        # ROC plot if probs present & binary
        if (y_prob is not None) and ensure_binary(y_train):
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            auc = roc_auc_score(y_test, y_prob)
            plt.figure(); plt.plot(fpr, tpr, label=f"AUC={auc:.3f}"); plt.plot([0,1],[0,1],"--"); plt.title(f"ROC - {name}"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend()
            plt.savefig(os.path.join(OUT_DIR, f"roc_{name}.png"), dpi=200, bbox_inches='tight'); plt.show(); plt.clf()
    except Exception as e:
        print(f"Model {name} failed:", e)

# -----------------------------
# 10) Ensembles (bagging, boosting) + Voting + Stacking
# -----------------------------
ens_models = {
    "RandomForest": RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE),
    "ExtraTrees": ExtraTreesClassifier(n_estimators=200, random_state=RANDOM_STATE),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=RANDOM_STATE),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state=RANDOM_STATE)
}
# Try to add XGBoost/LightGBM/CatBoost if installed
if HAVE_XGB:
    try:
        ens_models["XGBoost"] = xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=RANDOM_STATE)
    except Exception:
        pass
if HAVE_LGB:
    try:
        ens_models["LightGBM"] = lgb.LGBMClassifier(random_state=RANDOM_STATE)
    except Exception:
        pass
if HAVE_CAT:
    try:
        ens_models["CatBoost"] = cb.CatBoostClassifier(verbose=0, random_state=RANDOM_STATE)
    except Exception:
        pass

for name, mdl in ens_models.items():
    try:
        mdl.fit(X_train, y_train)
        fitted[name] = mdl
        y_pred = mdl.predict(X_test)
        y_prob = mdl.predict_proba(X_test)[:,1] if hasattr(mdl, "predict_proba") and ensure_binary(y_train) else None
        res = metrics_summary(y_test, y_pred, y_prob)
        results[name] = res
        print(f"\n[{name}] metrics:", res)
    except Exception as e:
        print(f"Ensemble {name} failed:", e)

# Voting classifier
voters = []
for v in ["LogisticRegression", "RandomForest", "DecisionTree"]:
    if v in fitted:
        voters.append((v, fitted[v]))
if voters:
    try:
        voting = VotingClassifier(estimators=voters, voting='soft')
        voting.fit(X_train, y_train)
        fitted["Voting"] = voting
        y_pred = voting.predict(X_test)
        y_prob = voting.predict_proba(X_test)[:,1] if ensure_binary(y_train) else None
        results["Voting"] = metrics_summary(y_test, y_pred, y_prob)
        print("\n[Voting] metrics:", results["Voting"])
    except Exception as e:
        print("Voting failed:", e)

# Stacking classifier
stack_estimators = [(name, fitted[name]) for name in ("RandomForest", "ExtraTrees", "GradientBoosting") if name in fitted]
if not stack_estimators:
    stack_estimators = [(name, models[name]) for name in ("RandomForest", "ExtraTrees") if name in models]
if stack_estimators:
    try:
        stacking = StackingClassifier(estimators=stack_estimators, final_estimator=LogisticRegression(), cv=5, n_jobs=1)
        stacking.fit(X_train, y_train)
        fitted["Stacking"] = stacking
        y_pred = stacking.predict(X_test)
        y_prob = stacking.predict_proba(X_test)[:,1] if ensure_binary(y_train) else None
        results["Stacking"] = metrics_summary(y_test, y_pred, y_prob)
        print("\n[Stacking] metrics:", results["Stacking"])
    except Exception as e:
        print("Stacking failed:", e)

# -----------------------------
# 11) Simple Deep Learning (Keras MLP) with training history plot; fallback to sklearn MLP
# -----------------------------
if HAVE_KERAS:
    try:
        print("Training Keras MLP...")
        n_in = X_train.shape[1]
        model = keras.Sequential([
            layers.Input(shape=(n_in,)),
            layers.Dense(64, activation="relu"),
            layers.Dropout(0.2),
            layers.Dense(32, activation="relu"),
            layers.Dense(1, activation="sigmoid") if ensure_binary(y_train) else layers.Dense(len(np.unique(y_train)), activation="softmax")
        ])
        model.compile(optimizer="adam", loss="binary_crossentropy" if ensure_binary(y_train) else "sparse_categorical_crossentropy", metrics=["accuracy"])
        hist = model.fit(X_train.values, y_train, validation_split=0.1 if X_train.shape[0]>10 else 0.0, epochs=50, batch_size=16, verbose=0)
        # save metrics plot
        plt.figure(); plt.plot(hist.history['loss'], label='train_loss')
        if 'val_loss' in hist.history: plt.plot(hist.history['val_loss'], label='val_loss')
        plt.title("Keras MLP Loss"); plt.legend(); plt.savefig(os.path.join(OUT_DIR, "keras_mlp_loss.png")); plt.clf()
        plt.figure(); plt.plot(hist.history.get('accuracy', hist.history.get('acc', [])), label='train_acc')
        if 'val_accuracy' in hist.history: plt.plot(hist.history['val_accuracy'], label='val_acc')
        plt.title("Keras MLP Accuracy"); plt.legend(); plt.savefig(os.path.join(OUT_DIR, "keras_mlp_acc.png")); plt.clf()
        # evaluate
        eval_res = model.evaluate(X_test.values, y_test, verbose=0)
        print("Keras MLP evaluation:", eval_res)
        model.save(os.path.join(OUT_DIR, "keras_mlp.h5"))
        results["Keras_MLP"] = {"eval": eval_res}
    except Exception as e:
        print("Keras training failed, falling back to sklearn MLP:", e)
        HAVE_KERAS = False

if not HAVE_KERAS:
    try:
        from sklearn.neural_network import MLPClassifier
        mlp = MLPClassifier(hidden_layer_sizes=(64,32), max_iter=500, random_state=RANDOM_STATE)
        mlp.fit(X_train, y_train)
        fitted["Sklearn_MLP"] = mlp
        y_pred = mlp.predict(X_test)
        y_prob = mlp.predict_proba(X_test)[:,1] if hasattr(mlp, "predict_proba") and ensure_binary(y_train) else None
        results["Sklearn_MLP"] = metrics_summary(y_test, y_pred, y_prob)
        print("\n[Sklearn_MLP] metrics:", results["Sklearn_MLP"])
    except Exception as e:
        print("Sklearn MLP also failed:", e)

# -----------------------------
# 12) SHAP explainability for final best model (use RandomForest if present)
# -----------------------------
best_model_for_shap = None
for cand in ("RandomForest", "ExtraTrees", "Stacking", "Voting", "LogisticRegression"):
    if cand in fitted:
        best_model_for_shap = fitted[cand]; break

if best_model_for_shap is None:
    # fallback to rf_shap or rf_perm if present
    try:
        best_model_for_shap = rf_shap
    except Exception:
        best_model_for_shap = None

if best_model_for_shap is not None and shap_available:
    try:
        explainer2 = shap.TreeExplainer(best_model_for_shap)
        shap_vals2 = explainer2.shap_values(X_test, check_additivity=False)
        # plot summary of test set
        shap.summary_plot(shap_vals2, X_test, feature_names=X_final.columns, show=False)
        plt.savefig(os.path.join(OUT_DIR, "shap_summary_test.png"), dpi=200, bbox_inches='tight'); plt.clf()
        print("Saved shap_summary_test.png")
    except Exception as e:
        print("SHAP explainability failed on chosen model:", e)
else:
    print("No suitable model for SHAP explainability or SHAP not available.")

# -----------------------------
# 13) Save results, models, and summary
# -----------------------------
print("Saving artifacts...")
# models (pickle)
for name, mdl in fitted.items():
    try:
        with open(os.path.join(OUT_DIR, f"model_{name}.pkl"), "wb") as f:
            pickle.dump(mdl, f)
    except Exception:
        pass

# results summary DataFrame
results_df = pd.DataFrame(results).T
results_df.to_csv(os.path.join(OUT_DIR, "model_results_summary.csv"))
print("Saved model_results_summary.csv")

# run summary
run_summary = {
    "timestamp": datetime.utcnow().isoformat() + "Z",
    "data_shape": df.shape,
    "selected_features": final_features,
    "models_trained": list(results.keys()),
    "outputs_folder": os.path.abspath(OUT_DIR)
}
with open(os.path.join(OUT_DIR, "run_summary.json"), "w") as f:
    json.dump(run_summary, f, indent=2)

print("All done. Artifacts in folder:", os.path.abspath(OUT_DIR))
print("You can inspect outputs/ for CSVs, PNGs, and saved models.")
```

```python
#!/usr/bin/env python3
"""
Comprehensive ML/DL pipeline for tabular dataset (UTI.csv)
Implements Preprocessing, Scaling, Imbalance handling, Correlation Analysis, SHAP, LIME,
Classical models, ensembles, stacking, and many deep-learning models including tabular
and image-style models (with thoughtful fallbacks).

Author: ChatGPT (GPT-5 Thinking mini)
"""

import os
import sys
import warnings
warnings.filterwarnings("ignore")
import json
import pickle
import subprocess
from datetime import datetime
from typing import List

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# SKLearn imports
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             confusion_matrix, roc_auc_score, roc_curve, classification_report)
from sklearn.utils import resample
from sklearn.inspection import permutation_importance

# Basic models
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import (RandomForestClassifier, ExtraTreesClassifier,
                              AdaBoostClassifier, GradientBoostingClassifier,
                              VotingClassifier, StackingClassifier)

# Optional heavy libraries — we try to import and set flags
def try_import(pkg_name, import_stmt):
    try:
        loc = {}
        exec(import_stmt, globals(), loc)
        return loc
    except Exception:
        return None

# imblearn SMOTE
try:
    from imblearn.over_sampling import SMOTE
    HAVE_SMOTE = True
except Exception:
    HAVE_SMOTE = False

# SHAP
try:
    import shap
    HAVE_SHAP = True
except Exception:
    HAVE_SHAP = False
    try:
        print("Attempting to pip install shap (optional)...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "shap"], stdout=subprocess.DEVNULL)
        import shap
        HAVE_SHAP = True
    except Exception:
        HAVE_SHAP = False

# LIME
try:
    from lime.lime_tabular import LimeTabularExplainer
    HAVE_LIME = True
except Exception:
    HAVE_LIME = False
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "lime"], stdout=subprocess.DEVNULL)
        from lime.lime_tabular import LimeTabularExplainer
        HAVE_LIME = True
    except Exception:
        HAVE_LIME = False

# TensorFlow/Keras
try:
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
    HAVE_KERAS = True
except Exception:
    HAVE_KERAS = False

# XGBoost, LightGBM, CatBoost
try:
    import xgboost as xgb
    HAVE_XGB = True
except Exception:
    HAVE_XGB = False

try:
    import lightgbm as lgb
    HAVE_LGB = True
except Exception:
    HAVE_LGB = False

try:
    import catboost as cb
    HAVE_CAT = True
except Exception:
    HAVE_CAT = False

# PyTorch & timm (for ViT via timm) — optional
try:
    import torch
    HAVE_TORCH = True
except Exception:
    HAVE_TORCH = False

try:
    import timm
    HAVE_TIMM = True
except Exception:
    HAVE_TIMM = False

# TabNet / TabTransformer / SAINT / FT-Transformer libraries (optional)
# Packages: pytorch_tabnet, pytorch_tabnet.TabNetClassifier, tab-transformer implementations etc.
try:
    from pytorch_tabnet.tab_model import TabNetClassifier, TabNetRegressor
    HAVE_TABNET = True
except Exception:
    HAVE_TABNET = False

# A simple flag group to choose heavy options
ENABLE_TABNET = False      # set True to attempt TabNet (requires pytorch_tabnet)
ENABLE_TABTRANSFORMER = False  # set True to attempt TabTransformer (requires custom libs)
ENABLE_VISION_MODELS = False   # set True to attempt image-style ViT/CNN using reshaped tabular data

# ------------------------------
# USER PARAMETERS (edit here)
# ------------------------------
DATA_PATH = "UTI.csv"   # your dataset path
TARGET_NAME = "uti"               # exact column name in CSV
SCALER = "standard"               # "standard" or "minmax" OR "both"
CORR_THRESHOLD = 0.85
TOP_K_SHAP = 10
TEST_SIZE = 0.2
RANDOM_STATE = 42
OUT_DIR = "outputs"
os.makedirs(OUT_DIR, exist_ok=True)

# How many instances to explain with LIME / SHAP local
NUM_EXPLAIN = 3

# If dataset is large, pairplot will be skipped
PAIRPLOT_MAX_ROWS = 500
PAIRPLOT_MAX_COLS = 6

# ------------------------------
# Utility functions
# ------------------------------
def save_fig(plt_obj, name):
    path = os.path.join(OUT_DIR, name)
    plt_obj.tight_layout()
    plt_obj.savefig(path, dpi=200, bbox_inches='tight')
    plt_obj.clf()
    return path

def ensure_binary(y):
    return len(np.unique(y)) == 2

def metrics_dict(y_true, y_pred, y_prob=None):
    avg = 'binary' if ensure_binary(y_true) else 'macro'
    out = {
        "accuracy": float(accuracy_score(y_true, y_pred)),
        "precision": float(precision_score(y_true, y_pred, average=avg, zero_division=0)),
        "recall": float(recall_score(y_true, y_pred, average=avg, zero_division=0)),
        "f1": float(f1_score(y_true, y_pred, average=avg, zero_division=0))
    }
    if (y_prob is not None) and ensure_binary(y_true):
        try:
            out["auc"] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            out["auc"] = None
    else:
        out["auc"] = None
    return out

def try_save_pickle(obj, name):
    try:
        with open(os.path.join(OUT_DIR, name), "wb") as f:
            pickle.dump(obj, f)
    except Exception as e:
        print("Could not save pickle", name, e)

# ------------------------------
# Load data
# ------------------------------
print("Loading dataset:", DATA_PATH)
if not os.path.exists(DATA_PATH):
    raise FileNotFoundError(f"{DATA_PATH} not found. Update DATA_PATH at top of script.")

df = pd.read_csv(DATA_PATH)
print("Shape:", df.shape)
print("Columns:", list(df.columns))

if TARGET_NAME not in df.columns:
    # attempt common names fallback
    candidates = [c for c in df.columns if c.lower() in ('target','label','class','diagnosis','uti','infection','result')]
    if candidates:
        TARGET_NAME = candidates[0]
        print("TARGET not found; using fallback:", TARGET_NAME)
    else:
        TARGET_NAME = df.columns[-1]
        print("TARGET not found; using last column:", TARGET_NAME)

print("Using TARGET:", TARGET_NAME)

# Quick EDA
print(df.head())
print(df.dtypes)
print("Missing per column:\n", df.isnull().sum())

# ------------------------------
# Preprocessing: missing values, encoding
# ------------------------------
# Drop columns with >60% missing
miss_frac = df.isnull().mean()
cols_drop = miss_frac[miss_frac > 0.6].index.tolist()
if cols_drop:
    print("Dropping cols with >60% missing:", cols_drop)
    df = df.drop(columns=cols_drop)

# Drop rows that are all NA
df = df.dropna(how='all').reset_index(drop=True)

# Separate X & y
X_raw = df.drop(columns=[TARGET_NAME]).copy()
y_raw = df[TARGET_NAME].copy()

# Numeric / categorical columns
num_cols = X_raw.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = X_raw.select_dtypes(exclude=[np.number]).columns.tolist()
print("Numeric cols:", num_cols)
print("Categorical cols:", cat_cols)

# Impute numeric with median, categorical with mode
if num_cols:
    X_raw[num_cols] = SimpleImputer(strategy="median").fit_transform(X_raw[num_cols])
if cat_cols:
    X_raw[cat_cols] = SimpleImputer(strategy="most_frequent").fit_transform(X_raw[cat_cols])
    # Label encode categorical features (safe for tree models; for DNN we will one-hot if needed)
    for c in cat_cols:
        X_raw[c] = LabelEncoder().fit_transform(X_raw[c].astype(str))

# Encode target if necessary
if y_raw.dtype == object or y_raw.dtype.name == 'category':
    y = LabelEncoder().fit_transform(y_raw.astype(str))
else:
    # try to cast to int if possible
    try:
        y = pd.Series(y_raw).astype(int).values
    except Exception:
        y = pd.Series(y_raw).values

# Save cleaned data
pd.concat([pd.DataFrame(X_raw), pd.Series(y, name=TARGET_NAME)], axis=1).to_csv(os.path.join(OUT_DIR, "data_cleaned.csv"), index=False)

# ------------------------------
# Scaling options
# ------------------------------
scalers = {}
scalers['standard'] = StandardScaler()
scalers['minmax'] = MinMaxScaler()

# If SCALER == "both", we'll produce both scaled datasets and run models on Standard by default
if SCALER not in ('standard', 'minmax', 'both'):
    SCALER = 'standard'
print("Using SCALER option:", SCALER)

# create scaled copy (we'll default to standard for modelling)
X_scaled_standard = pd.DataFrame(scalers['standard'].fit_transform(X_raw), columns=X_raw.columns)
X_scaled_minmax = pd.DataFrame(scalers['minmax'].fit_transform(X_raw), columns=X_raw.columns)
X_scaled = X_scaled_standard if SCALER in ('standard','both') else X_scaled_minmax
X_scaled.to_csv(os.path.join(OUT_DIR, "X_scaled.csv"), index=False)

# ------------------------------
# Imbalance handling: SMOTE or random oversample
# ------------------------------
vals, counts = np.unique(y, return_counts=True)
print("Class distribution before balancing:", dict(zip(vals, counts)))
X_bal = X_scaled.copy()
y_bal = y.copy()

if len(counts) > 1 and counts.max()/counts.min() > 1.5:
    print("Significant imbalance detected.")
    if HAVE_SMOTE:
        try:
            sm = SMOTE(random_state=RANDOM_STATE)
            X_bal, y_bal = sm.fit_resample(X_bal, y_bal)
            print("SMOTE applied.")
        except Exception as e:
            print("SMOTE failed:", e)
            HAVE_SMOTE = False
    if not HAVE_SMOTE:
        # random oversample minority classes to majority class
        df_comb = pd.concat([X_bal, pd.Series(y_bal, name=TARGET_NAME)], axis=1)
        majority = df_comb[TARGET_NAME].value_counts().max()
        parts = []
        for cls, grp in df_comb.groupby(TARGET_NAME):
            if len(grp) < majority:
                parts.append(resample(grp, replace=True, n_samples=majority, random_state=RANDOM_STATE))
            else:
                parts.append(grp)
        df_bal = pd.concat(parts).sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)
        y_bal = df_bal[TARGET_NAME].values
        X_bal = df_bal.drop(columns=[TARGET_NAME])
        print("Random oversampling applied.")
else:
    print("No major imbalance or ratio <= 1.5; skipping oversampling.")

print("Class distribution after balancing:", dict(pd.Series(y_bal).value_counts()))
pd.concat([X_bal, pd.Series(y_bal, name=TARGET_NAME)], axis=1).to_csv(os.path.join(OUT_DIR, "data_balanced.csv"), index=False)

# ------------------------------
# Correlation analysis + heatmap + remove highly correlated features
# ------------------------------
print("Computing correlation matrix and removing features with |corr| > {:.2f}".format(CORR_THRESHOLD))
corr = X_bal.corr().abs()
corr.to_csv(os.path.join(OUT_DIR, "correlation_matrix.csv"))
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Correlation heatmap")
plt.savefig(os.path.join(OUT_DIR, "correlation_heatmap.png"), dpi=200, bbox_inches='tight')
plt.clf()

# remove highly correlated columns (keep one of each correlated pair)
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > CORR_THRESHOLD)]
print("Dropping correlated features:", to_drop)
X_uncorr = X_bal.drop(columns=to_drop, errors='ignore')
X_uncorr.to_csv(os.path.join(OUT_DIR, "X_uncorr.csv"), index=False)

# ------------------------------
# Feature importance: SHAP or permutation importance fallback
# ------------------------------
print("Computing feature importance via SHAP (TreeExplainer) or permutation fallback.")
selected_features = []
shap_data = None
if HAVE_SHAP:
    try:
        rf = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=1)
        rf.fit(X_uncorr, y_bal)
        explainer = shap.TreeExplainer(rf)
        shap_values = explainer.shap_values(X_uncorr, check_additivity=False)
        if isinstance(shap_values, list) and len(shap_values) == 2:
            shap_arr = shap_values[1]
        else:
            shap_arr = shap_values
        shap_mean_abs = np.abs(shap_arr).mean(axis=0)
        shap_ser = pd.Series(shap_mean_abs, index=X_uncorr.columns).sort_values(ascending=False)
        shap_ser.to_csv(os.path.join(OUT_DIR, "shap_mean_abs.csv"))
        selected_features = list(shap_ser.head(TOP_K_SHAP).index)
        shap_data = (explainer, shap_arr)
        print("Top SHAP features:", selected_features)
    except Exception as e:
        print("SHAP failed:", e)
        HAVE_SHAP = False

if not HAVE_SHAP:
    print("Using permutation importance fallback")
    rf_p = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=1)
    rf_p.fit(X_uncorr, y_bal)
    perm = permutation_importance(rf_p, X_uncorr, y_bal, n_repeats=10, random_state=RANDOM_STATE, n_jobs=1)
    perm_ser = pd.Series(perm.importances_mean, index=X_uncorr.columns).sort_values(ascending=False)
    perm_ser.to_csv(os.path.join(OUT_DIR, "permutation_importances.csv"))
    selected_features = list(perm_ser.head(TOP_K_SHAP).index)
    print("Top permutation features:", selected_features)

# Final selected features after SHAP/permutation and correlation removal
final_features = [f for f in selected_features if f in X_uncorr.columns]
if not final_features:
    raise RuntimeError("No features selected. Check dataset or TOP_K_SHAP.")
print("Final features selected:", final_features)
pd.Series(final_features).to_csv(os.path.join(OUT_DIR, "selected_features.csv"), index=False)
X_final = X_uncorr[final_features].copy()
X_final.to_csv(os.path.join(OUT_DIR, "X_final_selected.csv"), index=False)

# ------------------------------
# SHAP visualizations (summary + bar + dependence top 3)
# ------------------------------
if HAVE_SHAP and shap_data is not None:
    explainer, shap_arr = shap_data
    try:
        print("Saving SHAP summary (beeswarm)...")
        shap.summary_plot(shap_arr, X_uncorr, show=False)
        plt.savefig(os.path.join(OUT_DIR, "shap_summary.png"), dpi=200, bbox_inches='tight'); plt.clf()
        print("Saved shap_summary.png")
    except Exception as e:
        print("Could not save SHAP summary:", e)
    try:
        print("Saving SHAP bar plot...")
        shap.summary_plot(shap_arr, X_uncorr, plot_type="bar", show=False)
        plt.savefig(os.path.join(OUT_DIR, "shap_bar.png"), dpi=200, bbox_inches='tight'); plt.clf()
        print("Saved shap_bar.png")
    except Exception as e:
        print("Could not save SHAP bar:", e)
    # dependence plots for top 3 final features
    for feat in final_features[:3]:
        try:
            shap.dependence_plot(feat, shap_arr, X_uncorr, show=False)
            plt.savefig(os.path.join(OUT_DIR, f"shap_dependence_{feat}.png"), dpi=200, bbox_inches='tight'); plt.clf()
            print("Saved shap_dependence for", feat)
        except Exception as e:
            print("Dependence plot failed for", feat, e)
else:
    print("SHAP not available; saved permutation importances instead.")

# ------------------------------
# Visualizations: histogram, boxplot, scatter, bar, pie, violin, pairplot conditional
# ------------------------------
print("Generating visualizations for the final features")
# Histogram of first selected feature
f1 = final_features[0]
plt.figure(figsize=(6,4)); sns.histplot(X_final[f1], bins=20, kde=True); plt.title(f"Histogram - {f1}")
plt.savefig(os.path.join(OUT_DIR, f"hist_{f1}.png"), dpi=200, bbox_inches='tight'); plt.clf()

# Boxplot of second feature if exists
if len(final_features) > 1:
    f2 = final_features[1]
    plt.figure(figsize=(6,4)); sns.boxplot(x=X_final[f2]); plt.title(f"Boxplot - {f2}")
    plt.savefig(os.path.join(OUT_DIR, f"box_{f2}.png"), dpi=200, bbox_inches='tight'); plt.clf()

# Scatter top2
if len(final_features) > 1:
    plt.figure(figsize=(6,4)); plt.scatter(X_final[final_features[0]], X_final[final_features[1]], alpha=0.7)
    plt.xlabel(final_features[0]); plt.ylabel(final_features[1]); plt.title("Scatter: Top2"); plt.savefig(os.path.join(OUT_DIR, "scatter_top2.png"), dpi=200, bbox_inches='tight'); plt.clf()

# Class bar and pie
vals2, cnts2 = np.unique(y_bal, return_counts=True)
plt.figure(figsize=(6,4)); plt.bar([str(v) for v in vals2], cnts2); plt.title("Class counts (after balancing)")
plt.savefig(os.path.join(OUT_DIR, "bar_class_counts.png"), dpi=200, bbox_inches='tight'); plt.clf()

plt.figure(figsize=(5,5)); plt.pie(cnts2, labels=[str(v) for v in vals2], autopct="%1.1f%%"); plt.title("Class distribution (%)")
plt.savefig(os.path.join(OUT_DIR, "pie_class_dist.png"), dpi=200, bbox_inches='tight'); plt.clf()

# Violin for up to 6 features
plt.figure(figsize=(10,6)); sns.violinplot(data=X_final.iloc[:, :min(6, X_final.shape[1])], inner="box"); plt.title("Violin: Top features")
plt.savefig(os.path.join(OUT_DIR, "violin_top_features.png"), dpi=200, bbox_inches='tight'); plt.clf()

# Pairplot if small
if X_final.shape[0] <= PAIRPLOT_MAX_ROWS and X_final.shape[1] <= PAIRPLOT_MAX_COLS:
    try:
        sns.pairplot(pd.concat([X_final, pd.Series(y_bal, name=TARGET_NAME)], axis=1), hue=TARGET_NAME, diag_kind="kde")
        plt.savefig(os.path.join(OUT_DIR, "pairplot_selected.png"), dpi=200, bbox_inches='tight'); plt.clf()
    except Exception as e:
        print("Pairplot failed:", e)
else:
    print("Pairplot skipped (dataset too large or too many features)")

# ------------------------------
# Train/test split on selected features
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(X_final, y_bal, test_size=TEST_SIZE, stratify=y_bal if len(np.unique(y_bal))>1 else None, random_state=RANDOM_STATE)
print("Train/Test shapes:", X_train.shape, X_test.shape)

# ------------------------------
# Classical ML models: fit & evaluate
# ------------------------------
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    "DecisionTree": DecisionTreeClassifier(random_state=RANDOM_STATE),
    "KNN": KNeighborsClassifier(),
    "GaussianNB": GaussianNB(),
    "SVM": SVC(probability=True, random_state=RANDOM_STATE)
}
results = {}
fitted_models = {}

for name, mdl in models.items():
    try:
        mdl.fit(X_train, y_train)
        fitted_models[name] = mdl
        y_pred = mdl.predict(X_test)
        y_prob = None
        if hasattr(mdl, "predict_proba") and ensure_binary(y_train):
            try: y_prob = mdl.predict_proba(X_test)[:,1]
            except: y_prob = None
        elif hasattr(mdl, "decision_function"):
            try: y_prob = mdl.decision_function(X_test)
            except: y_prob = None
        metrics = metrics_dict(y_test, y_pred, y_prob)
        results[name] = metrics
        print(f"\n{name} metrics:", metrics)
        print(classification_report(y_test, y_pred, zero_division=0))
        # confusion matrix
        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(4,3)); sns.heatmap(cm, annot=True, fmt="d"); plt.title(f"Confusion - {name}")
        plt.savefig(os.path.join(OUT_DIR, f"confusion_{name}.png"), dpi=200, bbox_inches='tight'); plt.clf()
        # ROC
        if (y_prob is not None) and ensure_binary(y_train):
            fpr, tpr, _ = roc_curve(y_test, y_prob)
            auc = roc_auc_score(y_test, y_prob)
            plt.figure(); plt.plot(fpr, tpr, label=f"AUC={auc:.3f}"); plt.plot([0,1],[0,1],"--"); plt.title(f"ROC - {name}"); plt.legend()
            plt.savefig(os.path.join(OUT_DIR, f"roc_{name}.png"), dpi=200, bbox_inches='tight'); plt.clf()
    except Exception as e:
        print(f"Model {name} failed:", e)

# ------------------------------
# Ensembles: RandomForest, ExtraTrees, AdaBoost, GradientBoosting, plus optional XGBoost/LGB/CatBoost
# ------------------------------
ens_models = {
    "RandomForest": RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE),
    "ExtraTrees": ExtraTreesClassifier(n_estimators=200, random_state=RANDOM_STATE),
    "AdaBoost": AdaBoostClassifier(n_estimators=100, random_state=RANDOM_STATE),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state=RANDOM_STATE)
}
if HAVE_XGB:
    try: ens_models["XGBoost"] = xgb.XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=RANDOM_STATE)
    except: pass
if HAVE_LGB:
    try: ens_models["LightGBM"] = lgb.LGBMClassifier(random_state=RANDOM_STATE)
    except: pass
if HAVE_CAT:
    try: ens_models["CatBoost"] = cb.CatBoostClassifier(verbose=0, random_state=RANDOM_STATE)
    except: pass

for name, mdl in ens_models.items():
    try:
        mdl.fit(X_train, y_train)
        fitted_models[name] = mdl
        y_pred = mdl.predict(X_test)
        y_prob = mdl.predict_proba(X_test)[:,1] if hasattr(mdl, "predict_proba") and ensure_binary(y_train) else None
        metrics = metrics_dict(y_test, y_pred, y_prob)
        results[name] = metrics
        print(f"\n{name} metrics:", metrics)
    except Exception as e:
        print(f"Ensemble {name} failed:", e)

# Voting classifier (soft) when possible
voters = []
for cand in ("LogisticRegression","RandomForest","DecisionTree"):
    if cand in fitted_models:
        voters.append((cand, fitted_models[cand]))
if voters:
    try:
        voting = VotingClassifier(estimators=voters, voting='soft')
        voting.fit(X_train, y_train)
        fitted_models["Voting"] = voting
        y_pred = voting.predict(X_test)
        y_prob = voting.predict_proba(X_test)[:,1] if ensure_binary(y_train) else None
        results["Voting"] = metrics_dict(y_test, y_pred, y_prob)
        print("\nVoting metrics:", results["Voting"])
    except Exception as e:
        print("Voting failed:", e)

# Stacking
stack_estimators = [(n, fitted_models[n]) for n in ("RandomForest","ExtraTrees","GradientBoosting") if n in fitted_models]
if not stack_estimators:
    stack_estimators = [("rf", RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE))]
try:
    stacking = StackingClassifier(estimators=stack_estimators, final_estimator=LogisticRegression(), cv=5, n_jobs=1)
    stacking.fit(X_train, y_train)
    fitted_models["Stacking"] = stacking
    y_pred = stacking.predict(X_test)
    y_prob = stacking.predict_proba(X_test)[:,1] if ensure_binary(y_train) else None
    results["Stacking"] = metrics_dict(y_test, y_pred, y_prob)
    print("\nStacking metrics:", results["Stacking"])
except Exception as e:
    print("Stacking failed:", e)

# ------------------------------
# Deep Learning: many variants (DNN, 1D-CNN, LSTM, GRU, TabNet (optional), TabTransformer (optional), ViT/CNN on reshaped tabular images (optional))
# Note: heavy — enable/disable with flags at top
# ------------------------------
dl_results = {}
if HAVE_KERAS:
    try:
        print("Training Keras DNN (MLP)...")
        n_in = X_train.shape[1]
        model = keras.Sequential([
            layers.Input(shape=(n_in,)),
            layers.Dense(128, activation="relu"),
            layers.Dropout(0.2),
            layers.Dense(64, activation="relu"),
            layers.Dense(32, activation="relu"),
            layers.Dense(1, activation="sigmoid") if ensure_binary(y_train) else layers.Dense(len(np.unique(y_train)), activation="softmax")
        ])
        model.compile(optimizer="adam", loss="binary_crossentropy" if ensure_binary(y_train) else "sparse_categorical_crossentropy", metrics=["accuracy"])
        hist = model.fit(X_train.values, y_train, validation_split=0.1 if X_train.shape[0]>10 else 0.0, epochs=30, batch_size=16, verbose=0)
        # save training curves
        plt.figure(); plt.plot(hist.history['loss'], label='train_loss')
        if 'val_loss' in hist.history: plt.plot(hist.history['val_loss'], label='val_loss')
        plt.legend(); plt.title("DNN Loss"); plt.savefig(os.path.join(OUT_DIR, "dnn_loss.png")); plt.clf()
        plt.figure(); plt.plot(hist.history.get('accuracy', hist.history.get('acc', [])), label='train_acc')
        if 'val_accuracy' in hist.history: plt.plot(hist.history['val_accuracy'], label='val_acc')
        plt.legend(); plt.title("DNN Acc"); plt.savefig(os.path.join(OUT_DIR, "dnn_acc.png")); plt.clf()
        # evaluate
        eval_res = model.evaluate(X_test.values, y_test, verbose=0)
        dl_results["DNN_Keras"] = eval_res
        model.save(os.path.join(OUT_DIR, "dnn_keras.h5"))
        fitted_models["DNN_Keras"] = model
        print("DNN_Keras eval:", eval_res)
    except Exception as e:
        print("Keras DNN failed:", e)
else:
    print("Keras not available: skipping Keras DNN")

# 1D-CNN on tabular as sequence (reshape features -> channels=1)
if HAVE_KERAS:
    try:
        print("Training simple 1D-CNN on tabular features (reshape to [samples, features, 1])")
        n_in = X_train.shape[1]
        model_cnn = keras.Sequential([
            layers.Reshape((n_in, 1), input_shape=(n_in,)),
            layers.Conv1D(64, kernel_size=2, activation='relu'),
            layers.MaxPooling1D(pool_size=2),
            layers.Flatten(),
            layers.Dense(64, activation='relu'),
            layers.Dense(1, activation='sigmoid') if ensure_binary(y_train) else layers.Dense(len(np.unique(y_train)), activation='softmax')
        ])
        model_cnn.compile(optimizer='adam', loss='binary_crossentropy' if ensure_binary(y_train) else 'sparse_categorical_crossentropy', metrics=['accuracy'])
        hist_cnn = model_cnn.fit(X_train.values, y_train, validation_split=0.1 if X_train.shape[0]>10 else 0.0, epochs=20, batch_size=16, verbose=0)
        model_cnn.save(os.path.join(OUT_DIR, "cnn1d_keras.h5"))
        fitted_models["CNN1D_Keras"] = model_cnn
        dl_results["CNN1D_Keras"] = model_cnn.evaluate(X_test.values, y_test, verbose=0)
        print("CNN1D eval:", dl_results["CNN1D_Keras"])
    except Exception as e:
        print("1D-CNN failed:", e)

# LSTM / GRU (sequence modeling) — interpret features as sequence
if HAVE_KERAS:
    try:
        print("Training LSTM on feature sequence (tabular as seq)...")
        n_in = X_train.shape[1]
        model_lstm = keras.Sequential([
            layers.Reshape((n_in, 1), input_shape=(n_in,)),
            layers.LSTM(64),
            layers.Dense(32, activation='relu'),
            layers.Dense(1, activation='sigmoid') if ensure_binary(y_train) else layers.Dense(len(np.unique(y_train)), activation='softmax')
        ])
        model_lstm.compile(optimizer='adam', loss='binary_crossentropy' if ensure_binary(y_train) else 'sparse_categorical_crossentropy', metrics=['accuracy'])
        hist_lstm = model_lstm.fit(X_train.values, y_train, validation_split=0.1 if X_train.shape[0]>10 else 0.0, epochs=20, batch_size=16, verbose=0)
        model_lstm.save(os.path.join(OUT_DIR, "lstm_keras.h5"))
        fitted_models["LSTM_Keras"] = model_lstm
        dl_results["LSTM_Keras"] = model_lstm.evaluate(X_test.values, y_test, verbose=0)
        print("LSTM eval:", dl_results["LSTM_Keras"])
    except Exception as e:
        print("LSTM failed:", e)

# GRU
if HAVE_KERAS:
    try:
        print("Training GRU on feature sequence (tabular as seq)...")
        n_in = X_train.shape[1]
        model_gru = keras.Sequential([
            layers.Reshape((n_in, 1), input_shape=(n_in,)),
            layers.GRU(64),
            layers.Dense(32, activation='relu'),
            layers.Dense(1, activation='sigmoid') if ensure_binary(y_train) else layers.Dense(len(np.unique(y_train)), activation='softmax')
        ])
        model_gru.compile(optimizer='adam', loss='binary_crossentropy' if ensure_binary(y_train) else 'sparse_categorical_crossentropy', metrics=['accuracy'])
        hist_gru = model_gru.fit(X_train.values, y_train, validation_split=0.1 if X_train.shape[0]>10 else 0.0, epochs=20, batch_size=16, verbose=0)
        model_gru.save(os.path.join(OUT_DIR, "gru_keras.h5"))
        fitted_models["GRU_Keras"] = model_gru
        dl_results["GRU_Keras"] = model_gru.evaluate(X_test.values, y_test, verbose=0)
        print("GRU eval:", dl_results["GRU_Keras"])
    except Exception as e:
        print("GRU failed:", e)

# Optional TabNet (if installed and enabled)
if ENABLE_TABNET and HAVE_TABNET:
    try:
        print("Training TabNet (pytorch_tabnet)...")
        tabnet = TabNetClassifier()
        tabnet.fit(X_train.values, y_train, eval_set=[(X_test.values, y_test)], max_epochs=100, patience=10, batch_size=256, virtual_batch_size=128)
        fitted_models["TabNet"] = tabnet
        try_save_pickle(tabnet, "model_TabNet.pkl")
    except Exception as e:
        print("TabNet failed:", e)

# Optional TabTransformer / other advanced tabular DL can be added similarly if libraries are present

# ------------------------------
# Generate LIME local explanations (option A: LIME local only)
# ------------------------------
if HAVE_LIME:
    try:
        print("Generating LIME local explanations for up to", NUM_EXPLAIN, "test instances")
        # choose a model for explanation: prefer RandomForest or Voting or Stacking
        lime_model = None
        for cand in ("RandomForest","Voting","Stacking","LogisticRegression"):
            if cand in fitted_models:
                lime_model = fitted_models[cand]; break
        if lime_model is None:
            lime_model = rf if 'rf' in locals() else None
        if lime_model is None:
            print("No suitable model found for LIME explanations; skipping.")
        else:
            explainer = LimeTabularExplainer(X_train.values, feature_names=X_final.columns.tolist(),
                                             class_names=[str(c) for c in np.unique(y_train)],
                                             mode='classification' if ensure_binary(y_train) else 'regression',
                                             random_state=RANDOM_STATE)
            n_explain = min(NUM_EXPLAIN, X_test.shape[0])
            for i in range(n_explain):
                inst = X_test.iloc[i].values
                explain_obj = explainer.explain_instance(inst, lime_model.predict_proba if ensure_binary(y_train) else lime_model.predict, num_features=min(10, X_final.shape[1]))
                html_path = os.path.join(OUT_DIR, f"lime_local_test_{i}.html")
                explain_obj.save_to_file(html_path)
                print("Saved LIME explanation:", html_path)
    except Exception as e:
        print("LIME failed:", e)
else:
    print("LIME not installed — skip LIME local explanations.")

# ------------------------------
# SHAP local explanations (force plot) for a few samples (if SHAP available and TreeExplainer used earlier)
# ------------------------------
if HAVE_SHAP and shap_data is not None:
    try:
        explainer_global, shap_arr_global = shap_data
        # Save force plots as html for a few instances
        for idx in range(min(NUM_EXPLAIN, X_test.shape[0])):
            try:
                single = X_test.iloc[[idx]]
                sv = explainer_global.shap_values(single, check_additivity=False)
                # force_plot HTML
                html_fp = os.path.join(OUT_DIR, f"shap_force_test_{idx}.html")
                try:
                    # For binary we want class 1
                    if isinstance(sv, list) and len(sv) == 2:
                        shap.force_plot(explainer_global.expected_value[1], sv[1], single, feature_names=X_final.columns).save_html(html_fp)
                    else:
                        shap.force_plot(explainer_global.expected_value, sv, single, feature_names=X_final.columns).save_html(html_fp)
                    print("Saved SHAP force plot:", html_fp)
                except Exception:
                    # fallback to bar summary PNG
                    arr = np.abs((sv[1] if isinstance(sv, list) else sv)).mean(axis=0)
                    pd.Series(arr, index=X_final.columns).sort_values(ascending=False).head(10).plot.bar()
                    plt.title(f"SHAP abs mean (instance {idx})"); plt.savefig(os.path.join(OUT_DIR, f"shap_local_bar_{idx}.png")); plt.clf()
            except Exception as e:
                print("SHAP local failed for idx", idx, e)
    except Exception as e:
        print("SHAP local explanations failed:", e)
else:
    print("SHAP not available or shap_data missing — skip SHAP local explanations.")

# ------------------------------
# Save models and run summary
# ------------------------------
print("Saving fitted models (pickles) and summary files...")
for name, m in fitted_models.items():
    try:
        try_save_pickle(m, f"model_{name}.pkl")
    except Exception as e:
        print("Could not pickle:", name, e)

# Save results summary (metrics)
results_df = pd.DataFrame(results).T
results_df.to_csv(os.path.join(OUT_DIR, "model_results_summary.csv"))

run_summary = {
    "timestamp": datetime.utcnow().isoformat() + "Z",
    "data_shape": df.shape,
    "selected_features": final_features,
    "models_trained": list(results.keys()),
    "outputs": os.path.abspath(OUT_DIR),
    "notes": {
        "HAVE_SHAP": HAVE_SHAP,
        "HAVE_LIME": HAVE_LIME,
        "HAVE_SMOTE": HAVE_SMOTE,
        "HAVE_KERAS": HAVE_KERAS,
        "HAVE_XGB": HAVE_XGB,
        "HAVE_LGB": HAVE_LGB,
        "HAVE_CAT": HAVE_CAT
    }
}
with open(os.path.join(OUT_DIR, "run_summary.json"), "w") as f:
    json.dump(run_summary, f, indent=2)

print("Pipeline complete. Artifacts saved to:", os.path.abspath(OUT_DIR))
print("Key outputs: model_results_summary.csv, selected_features.csv, correlation_heatmap.png, shap_*.png/html (if SHAP), lime_*.html (if LIME), many plots (hist, box, scatter, violin).")
```
